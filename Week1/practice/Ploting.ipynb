```python
import numpy as np
x=np.arange(0,10,0.1)
y=x**2
y1=x**3
plt.xlabel('x')
plt.ylabel('y')
plt.title('Demo')
plt.plot(x,y,color='Blue',marker='o',linewidth=2,label='x^2')
plt.plot(x,y1,color='Blue',marker='o',linewidth=2,label='x^3')
plt.legend()
plt.axis([0,20,0,1250])
plt.grid()
plt.text(12,600,'Graph',fontsize=20)
plt.show()
```

```python
# pie graph 
sizes=[3,2,5,7]
labels=['A','B','C','D']
colors=['blue','purple','green','pink']
explode=[0.1,0,0,0]
plt.pie(sizes,colors=colors, explode=explode,labels=labels,autopct='%.3f%%',counterclock=False)
plt.title('Pie Graph')
plt.show()
```
